```python
import numpy as np
import matplotlib.pyplot as plt

def minimos_cuadrados(x, y):
    """
    Metodo de mínimos cuadrados para ajustar una recta a los datos
    
    ----
    Parametros

    x : array
        Datos de la variable independiente
    y : array
        Datos de la variable dependiente
    ----
    """

    n = len(x)
    suma_x = np.sum(x)
    suma_y = np.sum(y)

    suma_x_cuadrado = np.sum(x**2)

    suma_x_resultado_cuadrado = suma_x**2

    suma_xy = np.sum(x * y)

    intercepto = ( (suma_y * suma_x_cuadrado) - (suma_x * suma_xy) )/(n * suma_x_cuadrado - suma_x_resultado_cuadrado)

    pendiente = ( ( n * suma_xy ) - (suma_x * suma_y) ) / ( (n * suma_x_cuadrado) - suma_x_resultado_cuadrado)


    return intercepto, pendiente


x = np.array([0, 2, 3, 6, 7])
y = np.array([0.120, 0.153, 0.170, 0.225, 0.260])

# Ajuste de la recta

intercepto, pendiente = minimos_cuadrados(x, y)

print(f"Intercepto: {intercepto}")
print(f"Pendiente: {pendiente}")

# Graficar los datos y la recta ajustada

plt.scatter(x, y, color='blue', label='Datos')
x_line = np.linspace(min(x), max(x), 100)
y_line = intercepto + pendiente * x_line
plt.plot(x_line, y_line, color='red', label='Ajuste de la recta')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste de Mínimos Cuadrados')
plt.legend()




```
